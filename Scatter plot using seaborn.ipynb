```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:\\Users\\HP\\Downloads\\Prices.csv')

# regplot build a scatter plot using seaborn
# x and y are columns which will be depicted as axes in graph
# data refers to dataframe name
# fit-reg=true will show best fit reagion and line also
sns.regplot(x='area', y='price', data=df,fit_reg=False)
plt.show()
```
