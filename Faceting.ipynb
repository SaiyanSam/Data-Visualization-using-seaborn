```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:\\Users\\HP\\Downloads\\Salaries.csv')

##First we need to apply FacetGrid method and pas a column whose each value will plot different graphs
data = sns.FacetGrid(df, col='discipline')

##Then we apply map method and pass the type of plot and the column to be represented as an axis
data = data.map(sns.kdeplot, 'salary')
plt.show()
```
